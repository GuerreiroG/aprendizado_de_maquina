```python
dados = pd.read_csv('mercado.csv', header = None)

transacoes = []
for i in range(0, 10):
    transacoes.append([str(dados.values[i, j]) for j in range(0, 4)])

regras = apriori(transacoes, min_support=0.3, min_confidence=0.8, min_lift=2, min_length=2)

resultados = list(regras)

resultados2 = [list(x) for x in resultados]

resultado_formatado = []
for j in range(0, 3):
    resultado_formatado.append([list(x) for x in resultados2[j][2]])

resultado_formatado
```

## 2) Utilize a base de dados maior (mercado2) que apresenta o histórico de compras de um supermercado francês em 1 semana. Teste o seu algoritmo e responda as seguintes perguntas:

### a) Vamos supor que você quer extrair Regras da Associação para os produtos que são vendidos pelo menos quatro vezes ao dia, ou seja, 28 vezes por semana (4$\times$7). Adicione o valor de suporte de acordo com o problema.

Para adicionar o valor de suporte, você deve calcular o valor de 28 (quantidade de produtos vendidos) dividido pela quantidade total de produtos na base de dados.

Execute o algoritmo e responda qual é o resultado.

```python
base_mercado = pd.read_csv('mercado2.csv', header = None)
base_mercado
```

```python
qtd_compras = len(base_mercado)
valor_suporte = 28/qtd_compras

transacoes = []
for i in range(qtd_compras):
    transacoes.append([str(base_mercado.values[i, j]) for j in range(20)
                       if str(base_mercado.values[i, j]) != 'nan'])

regras = apriori(transacoes, min_support=valor_suporte)
resultados = list(regras)

print(f"{len(resultados)} resultados com valor de suporte de {valor_suporte}")
print("Resultados:", resultados)
```

### b) Adicione o valor de confiança para 0.2 e o lift para 3. Quantos registros foram retornados?
Resposta: 44

```python
regras = apriori(transacoes, min_support=valor_suporte, min_confidence=0.2 ,min_lift=3)
resultados = list(regras)

print(f"{len(resultados)} registros encontrados.")
```

### c) Adicione o código abaixo para poder visualizar melhor os dados. Faça comentário do que acontece em cada linha do código.

```python
# Inicialização da lista A
A = []
# Inicialização da lista B
B = []
# Inicialização da lista de suporte
suporte = []
# Inicialização de confiança
confianca = []
# Inicialização de lift
lift = []

# Declaração do for para iterar sobre o resultados
for resultado in resultados:
    # atribui-se a "s" o valor do suporte do resultado
    s = resultado[1]
    # atribui a result rules a lista de regras ordenadas associadas àquele resultado
    result_rules = resultado[2]

    # Declaração do for para iterar sobre as regras ordenadas dos resultados
    for result_rule in result_rules:
        # atribui à "a" os produtos que compõe o antecedente da regra.
        a = list(result_rule[0])
        # atribui à "b" os produtos que compõe o consequente da regra.
        b = list(result_rule[1])
        # atribui à "c" a confiança da regra
        c = result_rule[2]
        # atribuia à "l" o lift da regra
        l = result_rule[3]
        # adiciona à "A" a lista produtos que compõe o antecedente da regra
        A.append(a)
        # adiciona à "B" a lista produtos que compõe o consequente da regra
        B.append(b)
        # adiciona à lista "suporte" o valor de suporte do resultado
        suporte.append(s)
        # adiciona à lista "confiança" a confiança da regra
        confianca.append(c)
        # adiciona à lista "lift" o lift da regra
        lift.append(l)
# Cria um DataFrame do pandas com os valores dos antecedentes, consequentes, suporte, confiança e lift.
rules_df = pd.DataFrame({'A': A, 'B': B, 'suporte': suporte, 'confianca': confianca, 'lift': lift})
# Orderna o DataFrame por lift em ordem decrescente.
rules_df.sort_values(by='lift', ascending=False)
```
